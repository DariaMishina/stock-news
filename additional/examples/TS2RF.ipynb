Давайте переведем этот временной ряд в предлагаемый мной табличный вид. Признаками будет - код страны и результаты переворота за прошлые 6 лет. Например в 2020 году для страны YY строчка данных будет выглядеть так:

[результат за 2014, результат за 2015, результат за 2016, результат за 2017, результат за 2018, результат за 2019, код страны]

а целевая переменная == результат за 2020


```
df.query('country == "XX"').iloc[-7:-1].coup.to_list() + ["XX"]
```

А целевая переменная - так:

```
df.query('country == "XX"').iloc[-1].coup
```

например так

```
out = []
for cntr in df.country.unique():
  print(cntr)
  tmp = df.loc[df.query(f'country == "{cntr}"').index]
  for i in range(-6, 0):
    column = f'result{i}'
    tmp[column] = tmp.coup.shift(-i)
  out.append(tmp)
```

```
df_new = pd.concat(out)
df_new
```

мы видим, что у части наших строк есть NA-значения - для них не было значений за 6 лет назад. Не страшно, выкинем все строчки с пропущеными значениями


```
df_new.dropna(inplace=True)
df_new
```

Теперь мы можем свести задачу к логистической регресии, например. Или лучше к CatBoost - он тут, скорее всего, лучше зайдет.
